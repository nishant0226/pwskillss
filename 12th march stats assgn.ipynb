```python
'''Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.
Answer1:
'''
alpha =1-0.95
sample_mean = 50
standard_deviation = 5
critical=1.96 
n=100 #assumptions made of 100 samples 
import scipy.stats as st
import numpy as np
import math
upper_ci=sample_mean+((critical*(standard_deviation/(np.sqrt(n)))))
lower_ci=sample_mean-((critical*(standard_deviation/(np.sqrt(n)))))
print("Confidence interval is",lower_ci,"to",upper_ci)
```

```python
'''Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.
'''
## Chi square Goodness OF Fit
expected_data=[20,20,20,10,10,20]
observed_data=[10,15,20,30,5,20] # assumed values in such a way that the sum of expected_data and observed_data will be same
chisquare_test_statistics,p_value=st.chisquare(observed_data,expected_data)
significance=0.05
dof=5
critical_value=st.chi2.ppf(1-significance,dof)
if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
'''Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:
            Group A     Group B

Outcome 1     20          15
Outcome 2     10          25
Outcome 3     15          20


Interpret the results of the test.
Answer3:
To calculate chi-square statistic and p-value for a contingency table with the following data:

'''
import scipy.stats as st
import numpy as np
contingency_table=np.array([[20,15],[10,25],[15,20]])
chisquare_test_statistics,p_value,dof,expected=st.chi2_contingency(contingency_table)
print(f'Chi-square test statistics is {chisquare_test_statistics} and p-value is {p_value}')
```

```python
'''Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.
Answer4:'''
import numpy as np
n=500
observed_value=60
critical_value=1.96
siginificance_value=0.05
p_hat=observed_value/n
#what is p_hat? p_hat is the sample proportion of individuals in the population who smoke
lower_ci=p_hat-(critical_value*(np.sqrt((p_hat*(1-p_hat))/n)))
upper_ci=p_hat+(critical_value*(np.sqrt((p_hat*(1-p_hat))/n)))
print("Confidence interval is",lower_ci,"to",upper_ci)


```

```python
'''Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.
Answer6:'''
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df = 10

x = np.linspace(0, 30, 500)  # Generate x-values for the plot
y = stats.chi2.pdf(x, df)    # Calculate the chi-square probability density function values

plt.plot(x, y,  label='df = 10 hehe')  # Plot the chi-square distribution
plt.fill_between(x,  y, where=(x >= 15), color='black' ,alpha=0.5)  # Shade the area for chi-square statistic of 15
#assumption of 95% C.I so the alpha value is 0.5 
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution (df = 10)')
plt.legend()

plt.show()
```

```python
'''Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.
Answer8:'''
#H0= The coin is biased towards tail
#h1: The coin is not biased towards tail
import scipy.stats as st

observed_tails=45
expected_tails=50
observed_heads=55
expected_heads=50
observed_data=[observed_tails,observed_heads]
expected_data=[expected_tails,expected_heads]
chisquare_test_statistics,p_value=st.chisquare(observed_data,expected_data)
significance=0.05
dof=1
critical_value=st.chi2.ppf(1-significance,dof)
if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
'''Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.

Use a significance level of 0.05.
                Lung Cancer: Yes    Lung Cancer: No
Smoker              60                  140
Non-smoker          30                  170
'''
#h0: there is an association
#h1: there is not an association
smokers=[60,140]
non_smokers=[30,170]
chisquare_test_statistics,p_value=st.chisquare(smokers,non_smokers)
significance=0.05
dof=1
critical_value=st.norm.ppf(1-significance,dof)
print(critical_value)
if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
'''Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05
Answer11:'''
#H0: The population mean is significantly different from 70
#H1: The population mean is not significantly different from 70
import scipy.stats as st
n=30
sample_mean=72
sample_standard_deviation=10
population_mean=70
significance=0.05
standard_error=sample_standard_deviation/np.sqrt(n)
z_test_statistics=(sample_mean-population_mean)/standard_error
critical_value=st.norm.ppf(1-significance)
print(critical_value)
if z_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
    
```
