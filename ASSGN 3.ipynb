```python
'''5. Scenario: A study is conducted to compare the performance of two different teaching methods. 
The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.
'''
from scipy import stats as st
a=[85, 90, 92, 88, 91]
b=[82, 88, 90, 86, 87]
mean_a=np.mean(a)
mean_b=np.mean(b)
std_a=np.std(a)
std_b=np.std(b)
n1=len(a)
n2=len(b)
t_test=mean_a-mean_b/(np.sqrt((std_a**2/n1)+(std_b**2/n2)))
print("t-test value is:",t_test)
#assuming significance level to be 0.05
alpha=0.05
#degree of freedom is 
dof=n1+n2-2
critical_value=st.t.ppf(1-alpha,dof)
print("Critical value is:",critical_value)  
if abs(t_test)>critical_value:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
'''8. Scenario: A company wants to analyze the relationship between employee tenure and job satisfaction. The data collected is as follows:
   Employee Tenure (in years): [2, 3, 5, 4, 6, 2, 4]
   Job Satisfaction (on a scale of 1 to 10): [7, 8, 6, 9, 5, 7, 6]
   Perform a linear regression analysis to predict job satisfaction based on employee tenure.
'''
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
tenure=[2, 3, 5, 4, 6, 2, 4]
satisfaction=[7, 8, 6, 9, 5, 7, 6]
tenure=np.array(tenure).reshape(-1,1)
satisfaction=np.array(satisfaction).reshape(-1,1)
reg=LinearRegression()
reg.fit(tenure,satisfaction)
print("Intercept is:",reg.intercept_)
print("Coefficient is:",reg.coef_)
plt.title("Linear Regression")
plt.xlabel("Tenure")
plt.ylabel("Satisfaction")
plt.grid()
plt.scatter(tenure,satisfaction)
plt.plot(tenure,reg.predict(tenure))
plt.show()
```

```python
'''9. Scenario: A study is conducted to compare the effectiveness of two different medications. 
The recovery times of the patients in each group are as follows:
   Medication A: [10, 12, 14, 11, 13]
   Medication B: [15, 17, 16, 14, 18]
   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the
    mean recovery times between the two medications.
'''
import scipy.stats as st

med_a = [10, 12, 14, 11, 13]
med_b = [15, 17, 16, 14, 18]

f_test = st.f_oneway(med_a, med_b)
print("F-Test value is:",f_test)
#assuming significance level to be 0.05
if f_test.pvalue < 0.05:
    print("There is a significant difference in the mean recovery times between the two medications.")
else:
    print("There is no significant difference in the mean recovery times between the two medications.")

```

```python
'''11. Scenario: A quality control department wants to test the weight consistency of a product. 
The weights of a sample of products are as follows:
    [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]
    Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams.
'''
#H0: mean weight=10
#H1: mean weight!=10
weights=[10.2, 9.8, 10.0, 10.5, 10.3, 10.1]
mean=np.mean(weights)
std=np.std(weights)
n=len(weights)
t_test=(mean-10)/(std/np.sqrt(n))
print("t-test value is:",t_test)
#assuming significance level to be 0.05
alpha=0.05
#degree of freedom is
dof=n-1
critical_value=st.t.ppf(1-alpha,dof)
print("Critical value is:",critical_value)
if abs(t_test)>critical_value:
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
'''12. Scenario: A company wants to analyze the click-through rates of two different website designs. 
The number of clicks for each design is as follows:
    Design A: [100, 120, 110, 90, 95]
    Design B: [80, 85, 90, 95, 100]
    Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.
'''
a=[100, 120, 110, 90, 95]
b=[80, 85, 90, 95, 100]
from scipy.stats import chi2_contingency
chi2, p, dof, ex = chi2_contingency([a,b])
print("Chi2 value is:",chi2)
print("p value is:",p)
#assuming significance level to be 0.05
if p < 0.05:
    print("There is a significant difference in the click-through rates between the two designs.")
else:
    print("There is no significant difference in the click-through rates between the two designs.")
    
```

```python
'''13. Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:
    [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
    Calculate the 95% confidence interval for the population mean satisfaction score.
'''
data=[7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
# confidence interval is 95% so alpha(significance level) is 0.05
alpha=0.05
#degree of freedom is
dof=len(data)-1
#critical value is
critical_value=st.t.ppf(1-alpha/2,dof)
print("Critical value is:",critical_value)
mean=np.mean(data)
std=np.std(data)
n=len(data)
confidence_interval=(mean-critical_value*(std/np.sqrt(n)),mean+critical_value*(std/np.sqrt(n)))
print("95% confidence interval for the population mean satisfaction score is:",confidence_interval)
```

```python
'''15. Scenario: A study is conducted to compare the preferences of two groups of participants. 
The preferences are measured on a Likert scale from 1 to 5. The data collected is as follows:
    Group A: [4, 3, 5, 2, 4]
    Group B: [3, 2, 4, 3, 3]
    Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups.
'''
from scipy.stats import mannwhitneyu
group_a=[4, 3, 5, 2, 4]
group_b=[3, 2, 4, 3, 3]
stat, p = mannwhitneyu(group_a, group_b)
print("Mann-Whitney U test value is:",stat)
print("p value is:",p)
#assuming significance level to be 0.05
if p < 0.05:
    print("There is a significant difference in the median preferences between the two groups.")
else:
    print("There is no significant difference in the median preferences between the two groups.")
    
```

```python
'''17. Scenario: A study is conducted to compare the performance of three different machine learning algorithms. 
The accuracy scores for each algorithm are as follows:
    Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]
    Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]
    Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]
    Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.
'''
from scipy.stats import kruskal
a=[0.85, 0.80, 0.82, 0.87, 0.83]
b=[0.78, 0.82, 0.84, 0.80, 0.79]
c=[0.90, 0.88, 0.89, 0.86, 0.87]
stat, p = kruskal(a, b, c)
print("Kruskal-Wallis test value is:",stat)
print("p value is:",p)
#assuming significance level to be 0.05
if p < 0.05:
    print("There is a significant difference in the median accuracy scores between the algorithms.")
else:
    print("There is no significant difference in the median accuracy scores between the algorithms.")
    
```
