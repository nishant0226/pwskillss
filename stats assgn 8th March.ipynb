```python
'''Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.
import math

def normal_pdf(x, mean, std_dev):
    """Calculate the probability density function of a normal distribution at a given point x,
    with a given mean and standard deviation.
    
    Args:
        x (float): The point at which to evaluate the PDF.
        mean (float): The mean of the normal distribution.
        std_dev (float): The standard deviation of the normal distribution.
    
    Returns:
        float: The value of the PDF at the given point x.
    """
    coefficient = 1.0 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -0.5 * ((x - mean) / std_dev)**2
    return coefficient * math.exp(exponent)

'''
```

```python
'''Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.
The binomial distribution is a probability distribution that describes the number of successes in a fixed number of 
independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. Some of the key 
properties of the binomial distribution are:

The trials are independent of each other.
There are a fixed number of trials, denoted by n.
Each trial has only two possible outcomes, denoted by p and q, where p is the probability of success and q is 
the probability of failure, such that p + q = 1.
The probability of success p is constant for all trials.
The random variable of interest is the number of successes, denoted by X.
The mean of the distribution is n*p.
The variance of the distribution is n*p*q.
Two examples of events where binomial distribution can be applied are:

Flipping a coin: When you flip a coin, there are only two possible outcomes: heads or tails. If you flip the coin n times, 
the number of heads you get will follow a binomial distribution with p=0.5 (assuming the coin is fair) and q=0.5. 
For example, if you flip a coin 10 times, the number of heads you get could be modeled by a binomial distribution with n=10, p=0.5, and q=0.5.

Quality control: In a manufacturing process, a certain percentage of items may be defective. If you inspect n items and 
record the number of defective items, the number of defects could be modeled by a binomial distribution with p representing 
the probability of an item being defective. For example, if you inspect 100 items and the probability of an item being defective is 0.1, 
then the number of defective items could be modeled by a binomial distribution with n=100 and p=0.1.

Overall, the binomial distribution is widely used in statistical analysis, particularly in applications that involve counting 
the number of successes or failures in a fixed number of independent trials with a constant probability of success.
'''
```

```python
'''Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.'''
import math
def poisson_cdf(x, mean):
    """Calculate the cumulative distribution function of a Poisson distribution at a given point x,
    with a given mean.
    
    Args:
        x (int): The point at which to evaluate the CDF.
        mean (float): The mean of the Poisson distribution.
    
    Returns:
        float: The value of the CDF at the given point x.
    """
    cdf = 0
    for i in range(x + 1):
        cdf += (mean**i / math.factorial(i)) * math.exp(-mean)
    return cdf
poisson_cdf(5, 3)
```

```python
'''Q7. How Binomial distribution different from Poisson distribution?

Binomial distribution and Poisson distribution are two common probability distributions that are often used in statistical analysis.

Binomial distribution is used to model the probability of a certain number of successes in a fixed number of independent trials and where the
outcomes are discrete (i.e., whole numbers). It is a special case of the Bernoulli distribution,
where each trial has a constant probability of success. It has two parameters: the number of trials n and the probability 
of success p in each trial.

Poisson distribution is used to model the probability of a certain number of events occurring in a fixed interval of time or space and where
the value can take any continuous value wihtin a range. 
It has one parameter: the expected number of events mu.

Here are some key differences between Binomial and Poisson distributions:

Binomial distribution models discrete outcomes (i.e., whole numbers) while Poisson distribution models counts (i.e., 
the number of occurrences in a continuous interval).

Binomial distribution requires a fixed number of trials and a constant probability of success, while Poisson distribution requires 
only an expected rate of occurrence.

The Poisson distribution is a limiting case of the binomial distribution where the number of trials becomes very large and the 
probability of success becomes very small, while keeping the product of n * p constant and equal to the expected rate of occurrence mu. 
As a result, when the expected number of events is low and the number of trials is large, the Poisson distribution provides a good 
approximation to the binomial distribution.

The variance of a Binomial distribution is equal to n * p * (1 - p), while the variance of a Poisson distribution is equal to its mean mu. 
This means that as the number of trials n increases and the probability of success p decreases in the Binomial distribution, 
the variance becomes larger and closer to the mean, while in the Poisson distribution, the variance and mean are always equal.

Binomial distribution can be used to model events that have only two outcomes (e.g., success or failure), 
while Poisson distribution can be used to model events that can occur any number of times (e.g., the number of phone calls received in a day).

In summary, Binomial distribution is used to model discrete events with a fixed number of trials and a constant probability of success, 
while Poisson distribution is used to model the number of occurrences in a continuous interval with a known expected rate of occurrence.
'''
```

```python
'''Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

'''
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
s = np.random.poisson(5, 1000)
sns.histplot(s,  kde=True,  color = 'black')
plt.show()
print('Sample mean:', np.mean(s))
print('Sample variance:', np.var(s))
```
